## Softmax Regression (Multinomial Logistic Regression)

Implementation of Softmax Regression, also known as multimonial logistic regression.

```python
import torch 
import torch.nn as nn
import torch.nn.functional as F
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from torch.utils.data import Dataset, DataLoader
from sklearn.datasets import make_blobs
import matplotlib.pylab as plt
from matplotlib.lines import Line2D
import numpy as np
import os
```

```python
device = "cpu"
if torch.cuda.is_available():
  print("Cuda is Available!")
  device = torch.device("cuda:0")
```

```python
train_dataset = datasets.FashionMNIST(root="../../data/",
                                      train=True,
                                      download=True,
                                      transform=transforms.ToTensor())
test_dataset = datasets.FashionMNIST(root="../../data/",
                                      train=True,
                                      download=False,
                                      transform=transforms.ToTensor())
```

```python
n_features = 28*28
n_dataset = len(test_dataset)
```

```python
train_loader = DataLoader(train_dataset,batch_size=128,shuffle=True)
test_loader = DataLoader(test_dataset,batch_size=1000)
```

```python
figure = plt.figure(figsize=(10, 8))
cols, rows = 5, 5
for i in range(1, cols * rows + 1):
    sample_idx = torch.randint(len(train_dataset), size=(1,)).item()
    img, label = train_dataset[sample_idx]
    figure.add_subplot(rows, cols, i)
    plt.title(label)
    plt.axis("off")
    plt.imshow(img.squeeze(), cmap="gray")
plt.show()
```

```python
class SoftmaxRegression(nn.Module):
  #Constructor
    def __init__(self,n_features,n_classes):
        super(SoftmaxRegression,self).__init__()
        self.linear = nn.Linear(n_features,n_classes)
        #self.linear.weight.detach().zero_()
        #self.linear.bias.detach().zero_()
        torch.nn.init.uniform_(self.linear.weight)
        self.n_features = n_features
    
    def forward(self,x):
        x = self.linear(x)
        return x
    
```

```python
model = SoftmaxRegression(n_features,n_classes=10)
```

```python
optimizer = torch.optim.SGD(model.parameters(),lr=0.2)
loss = nn.CrossEntropyLoss()
```

```python
def train(model,optimizer,loss_criterion,train_loader,test_loader,epochs,size):
    LOSS = []
    ACC = []
    for epoch in range(epochs):
        cost = 0
        for x, y in train_loader:
            x = x.view(-1,model.n_features)
            #y = y.to(device)
            optimizer.zero_grad()
            y_hat = model(x)
            loss = loss_criterion(y_hat,y)
            loss.backward()
            optimizer.step()
            cost+=loss.data
        LOSS.append(cost)
        
        correct = 0
        for x_test,y_test in test_loader:
            x_test = x_test.view(-1,model.n_features)#.to(device)
            #y_test = y_test.to(device)
            y_hat = model(x_test)
            _, predicted_labels = torch.max(y_hat.data, 1)
            correct += (predicted_labels == y_test).sum().item()
        accuracy = correct/size
        ACC.append(accuracy)
    return LOSS, ACC
```

```python
LOSS,ACC = train(model,optimizer,loss,train_loader,test_loader,1000,n_dataset)
```

```python
# Plot the loss and accuracy

fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.plot(LOSS, color=color)
ax1.set_xlabel('epoch', color=color)
ax1.set_ylabel('Cost', color=color)
ax1.tick_params(axis='y', color=color)
    
ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('accuracy', color=color) 
ax2.set_xlabel('epoch', color=color)
ax2.plot( ACC, color=color)
ax2.tick_params(axis='y', color=color)
fig.tight_layout()
```
